```python
# outlier detection and removing using IQR
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
sns.boxplot(x= "CoapplicantIncome",data=dataset)
plt.show()
```

```python
sns.boxplot(x= "ApplicantIncome",data=dataset)
plt.show()
```

```python
sns.distplot(dataset["ApplicantIncome"])
```

```python
dataset.shape
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q1
```

```python
q3 = dataset["CoapplicantIncome"].quantile(0.75)
q3
```

```python
IQR = q3 - q1
IQR
```

```python
lower_bound = q1 - (1.5)* IQR
upper_bound = q3 + (1.5)* IQR
print("lower bound is" , lower_bound)
print("upper bound is" , upper_bound)
```

```python
new_dataset = dataset[dataset["CoapplicantIncome"]<=upper_bound]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_dataset)
```

```python
# outlier detection and removing using z-score 
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(x="CoapplicantIncome",data=dataset)
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
mean = dataset["CoapplicantIncome"].mean()
mean
```

```python
sd = dataset["CoapplicantIncome"].std()
sd
```

```python
min_range = mean - 3*sd
max_range = mean + 3*sd
print("min range is ", min_range)
print("max range is ", max_range)
```

```python
new_dataset = dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_dataset)
```

```python
z_score = (dataset["CoapplicantIncome"] - mean)/sd
z_score
```

```python
dataset["z_score"] = z_score
dataset
```

```python
dataset[dataset["z_score"]<3]
```
