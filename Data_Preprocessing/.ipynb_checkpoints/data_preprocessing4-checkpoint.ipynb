```python
# onhot encoding
```

```python
import pandas as pd
```

```python
dataset=pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"] = dataset["Gender"].fillna(dataset["Gender"].mode()[0])
```

```python
dataset["Married"] = dataset["Married"].fillna(dataset["Married"].mode()[0])
```

```python
dataset.isnull().sum()
```

```python
en_data = dataset[["Gender","Married"]]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male","Married_No","Married_Yes"])
```

```python
ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes"])
```

```python
# ordinal encoding
```

```python
df = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head(3)
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
```

```python
df
```

```python
# encoding my map
```

```python
ord_data1 =  {"s":0,"m":1,"l":2,"xl":3}
```

```python
df["Size"].map(ord_data1)
```

```python
df["Size_en_map"] = df["Size"].map(ord_data1)
```

```python
df
```
